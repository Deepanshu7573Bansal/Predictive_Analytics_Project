## Student Performance Dataset

```python
import pandas as pd
```

```python
StudentPerformance=pd.read_csv("StudentsPerformance.csv")
```

```python
StudentPerformance.head()
```

```python
StudentPerformance=StudentPerformance.drop(columns=["gender","race/ethnicity","parental level of education","lunch","test preparation course"])
```

```python
StudentPerformance['category']="Null"
```

```python
StudentPerformance.head()
```

```python
math=StudentPerformance['math score']
reading=StudentPerformance['reading score']
writing=StudentPerformance['writing score']

for i in range(StudentPerformance.shape[0]):
    if (math[i]+reading[i]+writing[i])>200:
        StudentPerformance['category'][i]='A'
    elif (math[i]+reading[i]+writing[i])>100 and (math[i]+reading[i]+writing[i])<=200:
        StudentPerformance['category'][i]='B'
    elif (math[i]+reading[i]+writing[i])<=100:
        StudentPerformance['category'][i]='C'
```

```python
StudentPerformance.head()
```

```python
import pickle
pickle.dump(StudentPerformance,open("StudentPerformance.pkl","wb"))
```
